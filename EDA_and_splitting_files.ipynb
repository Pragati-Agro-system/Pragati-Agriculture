```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r'A:\CODING\Python\Machine Learning\HackathonAdityaFarmer\agridata_csv_202110311352.csv')
data
```

```python
data=data.dropna()
data['commodity'] = data['commodity_name']
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['commodity_name'] = le.fit_transform(data['commodity_name'])
data['state'] = le.fit_transform(data['state'])
data['district'] = le.fit_transform(data['district'])
data['market'] = le.fit_transform(data['market'])
```

```python
data
```

```python
#top 10 occuring commodities
commodity_count = data['commodity'].value_counts().nlargest(10)
commodity_count
```

```python
# storing top 10 commodities in differnt csv files
for commodity in commodity_count.index:
    commodity_data = data[data['commodity'] == commodity]
    filename = f'{commodity}.csv'
    #commodity_data.to_csv(filename, index=False)
    commodity_data.to_csv(r'A:\CODING\Python\Machine Learning\HackathonAdityaFarmer\my code\{}.csv'.format(filename), index=False)
```
